```python
from sklearn.model_selection import train_test_split

# Assuming 'X' contains your feature data and 'y' contains the corresponding labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
## StandarScaler
from sklearn.preprocessing import StandardScaler

# Assuming 'X' contains your feature data
scaler = StandardScaler()

# Fit the scaler on your data and transform the data
X_train = scaler.fit_transform(X_train)
X_test =  scaler.transform(X_test)

```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train[0]
```

# MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression

# Create a LinearRegression model
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test,reg_pred)
```

```python
##residuals:
resi = y_test-reg_pred
resi
```

```python
# res plot
sns.displot(resi, kind='kde')
#it should be normally distributed
```

```python
 plt.scatter(reg_pred,resi)
 #uniform distribution
```

```python
from  sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

mse = mean_squared_error(y_test, reg_pred)
root_mse = np.sqrt(mse)
```

```python
mse
```

```python
 #R2 and adjusted R2
from sklearn.metrics import r2_score
import numpy as np
y_pred = reg_pred
# Assuming 'y_test' contains the true target values and 'y_pred' contains the predicted values
r2 = r2_score(y_test, y_pred)

# Calculate adjusted R2
n = len(y_test)  # Number of samples
p = X.shape[1]    # Number of features
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("R-squared:", r2)
print("Adjusted R-squared:", adj_r2)
```
